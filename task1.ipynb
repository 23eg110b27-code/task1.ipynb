```python
#15.Which services are most popular by transaction count?

service_counts = df['services'].value_counts()

service_counts.head()
```

```python
#16.For each service, which product type is purchased most frequently?

# Count transactions per product within each service
top_products = df.groupby(['services', 'products_used']).size().reset_index(name='count')

# For each service, get the product with the highest count
most_frequent_products = top_products.loc[top_products.groupby('services')['count'].idxmax()]

most_frequent_products
```

```python
#17.What is the average transaction amount per product type?

avg_per_product = df.groupby('products_used')['t_amt'].mean()
print(avg_per_product)
```

```python
#18.Are there specific services where customers spend significantly more?

avg_per_service = df.groupby('services')['t_amt'].mean().sort_values(ascending=False)
print(avg_per_service)
```

```python
#19.Which state has the highest total sales?

# Sum total sales per state
total_sales_per_state = df.groupby('state')['t_amt'].sum()

# Find the state with the highest total sales
top_state = total_sales_per_state.idxmax()
top_sales = total_sales_per_state.max()

print(f"The state with the highest total sales is {top_state} with ${top_sales:.2f}.")
```

```python
#20.Which city records the highest number of transactions?

# Count transactions per city
transactions_per_city = df['city'].value_counts()

# City with the highest number of transactions
top_city = transactions_per_city.idxmax()
top_count = transactions_per_city.max()

print(f"The city with the highest number of transactions is {top_city} with {top_count} transactions.")
```

```python
#21.What is the average spending per transaction in each state?

top_city = df['city'].value_counts().idxmax()
top_count = df['city'].value_counts().max()

print(f"The city with the highest number of transactions is {top_city} with {top_count} transactions.")
```

```python
#22.Are certain services more popular in specific states?

service_state_counts = df.groupby(['state', 'services']).size().unstack(fill_value=0)

print(service_state_counts)
```

```python
#23.Which states buy the most Outdoor Recreation products?
# Filter for Outdoor Recreation service
outdoor_df = df[df['services'] == 'Outdoor Recreation']

# Count total transactions per state
state_counts = outdoor_df['state'].value_counts()

print(state_counts)
```

```python
#24.Compare average spending between California and Texas customers.
# Calculate average spending for California and Texas
avg_spending = df[df['state'].isin(['California', 'Texas'])] \
                .groupby('state')['t_amt'].mean()

print(avg_spending)
```

```python
#25.Which quarter of the year saw the highest sales?

import pandas as pd

# Convert t_date to datetime
df['t_date'] = pd.to_datetime(df['t_date'])

# Extract the quarter for each transaction
df['quarter'] = df['t_date'].dt.quarter

# Sum total sales per quarter
quarter_sales = df.groupby('quarter')['t_amt'].sum()

# Identify the quarter with the highest sales
top_quarter = quarter_sales.idxmax()
top_sales = quarter_sales.max()

print(f"Quarter {top_quarter} saw the highest sales: ${top_sales:.2f}")
```

```python
#26.How does total sales vary month by month?

import pandas as pd

# Convert t_date to datetime
df['t_date'] = pd.to_datetime(df['t_date'])

# Sum sales per month
monthly_sales = df.groupby(df['t_date'].dt.month)['t_amt'].sum()

print(monthly_sales)
```

```python
#27.What is the total number of transactions per month?

import pandas as pd

# Convert t_date to datetime
df['t_date'] = pd.to_datetime(df['t_date'])

# Count transactions per month
monthly_transactions = df.groupby(df['t_date'].dt.month).size()

print(monthly_transactions)
```

```python
#28. is there a particular season when sports equipment sales spike?

import pandas as pd
# Convert transaction date to datetime
df['t_date'] = pd.to_datetime(df['t_date'])
df['month'] = df['t_date'].dt.month

# Map month to season
df['season'] = df['month'].apply(lambda m: 'Winter' if m in [12,1,2] 
                                  else 'Spring' if m in [3,4,5] 
                                  else 'Summer' if m in [6,7,8] 
                                  else 'Fall')
sports_services = ['Team Sports', 'Racquet Sports', 'Gymnastics', 'Exercise & Fitness']
sports_df = df[df['services'].isin(sports_services)]
season_sales = sports_df.groupby('season')['t_amt'].sum().sort_values(ascending=False)
print(season_sales)
```

```python
#29.How many transactions were done using credit?

# Count credit transactions
credit_count = df[df['t_details'] == 'credit'].shape[0]

print(f"Number of credit transactions: {credit_count}")
```

```python
#30.What is the total revenue generated from credit transactions?

# Filter credit transactions and sum the amounts
credit_revenue = df[df['t_details'] == 'credit']['t_amt'].sum()

print(f"Total revenue from credit transactions: ${credit_revenue:.2f}")
```

```python
#31.Is there a difference in average spending between credit and debit customers?

# Calculate average spending by payment type
avg_spending = df.groupby('t_details')['t_amt'].mean()

print(avg_spending)
```

```python
#33.Should more inventory be stocked for Exercise & Fitness products given their popularity?

# Count of transactions per service
df['services'].value_counts()
df.groupby('services')['t_amt'].sum().sort_values(ascending=False)
```

```python
#34.Are there product categories with high sales but low average transaction amounts?

# Calculate total sales and average transaction per service
service_stats = df.groupby('services')['t_amt'].agg(['sum', 'mean']).sort_values(by='sum', ascending=False)

print(service_stats)
```

```python
35.Are there underperforming service categories that need promotional offers?

# Count transactions and sum sales per service
service_performance = df.groupby('services')['t_amt'].agg(['count', 'sum']).sort_values(by='sum')

print(service_performance)
```
